# Loading data

```python
import pandas as pd

df = pd.read_csv('../data/raw/Datafiniti_Hotel_Reviews.csv')
```

```python
df
```

```python
df.columns
```

# Removing unnecessary columns

```python
columns_to_drop = [
    'id',
    'dateAdded',
    'dateUpdated',
    'address',
    'categories',
    'primaryCategories',
    'city',
    'country',
    'keys',
    'latitude',
    'longitude',
    'name',
    'postalCode',
    'province',
    'reviews.date',
    'reviews.dateSeen',
    'reviews.sourceURLs',
    'reviews.title',
    'reviews.userCity',
    'reviews.userProvince',
    'reviews.username',
    'sourceURLs',
    'websites'
]

df = df.drop(labels=columns_to_drop, axis=1)
```

```python
df
```

# Renaming columns

```python
df = df.rename(columns={'reviews.rating': 'numerical_review'})
df = df.rename(columns={'reviews.text': 'comments'})
```

```python
df
```

# Removing empty rows

```python
print('Data information:')
df.info(show_counts=True)
```

```python
df = df.dropna(subset=['comments', 'numerical_review'], how='any')
```

```python
print('Data information:')
df.info(show_counts=True)
```

```python
df
```

# Detecting reviews' languages

```python
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException


DetectorFactory.seed = 0

def safe_detect(text):
    try:
        clean_text = text.strip()
        if len(clean_text) < 3:
            return 'unknown'
        return detect(clean_text)
    except LangDetectException:
        return 'unknown'
```

```python
df.loc[:, 'language'] = df['comments'].apply(lambda x: safe_detect(x) if pd.notnull(x) else 'unknown')
```

```python
df['language'].unique()
```

## Counting number of reviews in english

```python
english_count = df['language'].value_counts().get('en', 0)
english_count
```

## Removing non english reviews

```python
df = df[df['language'] == 'en']
```

```python
df['language'].unique()
```

# Adding sentiment column

```python
from textblob import TextBlob

df.loc[:, 'sentiment'] = df['comments'].apply(lambda x: TextBlob(str(x)).sentiment.polarity if pd.notna(x) else None)
```

```python
df
```

# Adding column with length of review

```python
df['review_length'] = df['comments'].str.len()
```

```python
df
```

# Assigining labels

```python
def assign_label(row):
    return 1 if row['numerical_review'] <= 2 or row['sentiment'] < -0.3 else 0
```

```python
df.loc[:, 'label'] = df.apply(assign_label, axis=1)
```

```python
df
```

# Reducing the data set

```python
label_0_sample = df[df['label'] == 0].sample(n=200, random_state=42)
other_labels = df[df['label'] != 0]
df = pd.concat([label_0_sample, other_labels])
```

```python
label_1_sample = df[df['label'] == 1].sample(n=100, random_state=42)
other_labels = df[df['label'] != 1]
df = pd.concat([label_1_sample, other_labels])
```

```python
df
```

# Saving prepared data

```python
df.to_csv('../data/processed/hotel_reviews.csv', index=False)
```
