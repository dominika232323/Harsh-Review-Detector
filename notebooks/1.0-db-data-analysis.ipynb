# Cleaning data

## Dropping columns

### With more than 50% missing values

```python
df = df.loc[:, df.isnull().mean() < 0.5]
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

### Unnecessary columns

```python
df = df.copy()

columns_to_remove = [
    'listing_url',
    'scrape_id',
    'last_scraped',
    'source',
    'picture_url',
    'host_id',
    'host_url',
    'host_name',
    'host_since',
    'host_location',
    'host_about',
    'host_has_profile_pic',
    'host_thumbnail_url',
    'host_picture_url',
    'host_listings_count',
    'host_total_listings_count',
    'host_verifications',
    'host_has_profile_pic',
    'calendar_last_scraped'
]
df.drop(columns_to_remove, axis=1, inplace=True)
df.head()
```

```python
print('Data description:')
df.describe(include='all')
```

## Converting price to numeric

```python
df['price'].unique()
```

```python
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
df.info()
```

## Fixing missing values

### In property_type and room_type columns

```python
# List every combination of values property_type and room_type columns
df.groupby(['property_type', 'room_type'], dropna=False).size().reset_index().rename(columns={0: 'count'})
```

```python
import numpy as np


df['room_type'] = df['room_type'].fillna(
    df.groupby('property_type')['room_type']
      .transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
)

df['property_type'] = df['property_type'].fillna(
    df.groupby('room_type')['property_type']
      .transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
)
```

```python
df.groupby(['property_type', 'room_type'], dropna=False).size().reset_index().rename(columns={0: 'count'})
```

### In bathrooms column

```python
df.value_counts('bathrooms', sort=True, ascending=False, dropna=False)
```

```python
df.loc[:, 'bathrooms_text_num'] = df['bathrooms_text'].str.extract(r'(\d*\.?\d+)').astype(float)
df.loc[:, 'bathrooms'] = df['bathrooms'].fillna(df['bathrooms_text_num'])

df.drop(columns=['bathrooms_text_num', 'bathrooms_text'], inplace=True)
```

```python
df.value_counts('bathrooms', sort=True, ascending=False, dropna=False)
```

# Locations of listings

```python
import folium

df_valid_coords = df.dropna(subset=['latitude', 'longitude'])

map = folium.Map(location=[df_valid_coords['latitude'].mean(), df_valid_coords['longitude'].mean()], zoom_start=12)

for _, row in df_valid_coords.iterrows():
    folium.CircleMarker([row['latitude'], row['longitude']], radius=2).add_to(map)

map
```

# Distribution of each review type

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d', 'number_of_reviews_ly']])
plt.title("Distribution of Reviews")
plt.ylabel("Number of Reviews")
plt.show()
```

# Correlation heatmap of each review type

```python
plt.figure(figsize=(8, 6))
corr = df[['number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d', 'number_of_reviews_ly']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Between Review Metrics")
plt.show()
```

# Price distribution by room type

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='room_type', y='price')
plt.title("Price Distribution by Room Type")
plt.ylabel("Price ($)")
plt.xlabel("Room Type")
plt.yscale("log")
plt.tight_layout()
plt.show()
```

# Price by Neighborhood

```python
plt.figure(figsize=(10, 20))
avg_price_by_neigh = df.groupby('neighbourhood_cleansed')['price'].mean().sort_values(ascending=False)
sns.barplot(x=avg_price_by_neigh.values, y=avg_price_by_neigh.index)
plt.title("Ranking of average price by neighborhood")
plt.xlabel("Average Price ($)")
plt.ylabel("Neighborhood")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
avg_price_by_neigh = df.groupby('neighbourhood_cleansed')['price'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=avg_price_by_neigh.values, y=avg_price_by_neigh.index)
plt.title("Top 10 Expensive Neighborhoods (Average Price)")
plt.xlabel("Average Price ($)")
plt.ylabel("Neighborhood")
plt.tight_layout()
plt.show()
```
