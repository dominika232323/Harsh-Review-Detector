# Cleaning data

## Dropping columns

### With more than 50% missing values

```python
df = df.loc[:, df.isnull().mean() < 0.5]
```

```python
df.info()
```

### Unnecessary columns

```python
df = df.copy()

columns_to_remove = [
    'id',
    'listing_url',
    'scrape_id',
    'last_scraped',
    'source',
    'picture_url',
    'host_id',
    'host_url',
    'host_name',
    'host_since',
    'host_location',
    'host_about',
    'host_has_profile_pic',
    'host_thumbnail_url',
    'host_picture_url',
    'host_listings_count',
    'host_total_listings_count',
    'host_verifications',
    'host_has_profile_pic',
    'neighbourhood_cleansed',
    'neighbourhood_group_cleansed',
    'calendar_last_scraped'
]
df.drop(columns_to_remove, axis=1, inplace=True)
df.head()
```

```python
print('Data description:')
df.describe(include='all')
```

## Converting price to numeric

```python
df['price'].unique()
```

```python
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
df.info()
```

## Fixing missing values

### In property_type and room_type columns

```python
# List every combination of values property_type and room_type columns
df.groupby(['property_type', 'room_type'], dropna=False).size().reset_index().rename(columns={0: 'count'})
```

```python
import numpy as np


df['room_type'] = df['room_type'].fillna(
    df.groupby('property_type')['room_type']
      .transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
)

df['property_type'] = df['property_type'].fillna(
    df.groupby('room_type')['property_type']
      .transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
)
```

```python
df.groupby(['property_type', 'room_type'], dropna=False).size().reset_index().rename(columns={0: 'count'})
```

### In bathrooms column

```python
df.value_counts('bathrooms', sort=True, ascending=False, dropna=False)
```

```python
df.loc[:, 'bathrooms_text_num'] = df['bathrooms_text'].str.extract(r'(\d*\.?\d+)').astype(float)
df.loc[:, 'bathrooms'] = df['bathrooms'].fillna(df['bathrooms_text_num'])

df.drop(columns=['bathrooms_text_num', 'bathrooms_text'], inplace=True)
```

```python
df.value_counts('bathrooms', sort=True, ascending=False, dropna=False)
```

# Locations of listings

```python
import folium

df_valid_coords = df.dropna(subset=['latitude', 'longitude'])

map = folium.Map(location=[df_valid_coords['latitude'].mean(), df_valid_coords['longitude'].mean()], zoom_start=12)

for _, row in df_valid_coords.iterrows():
    folium.CircleMarker([row['latitude'], row['longitude']], radius=2).add_to(map)

map
```
