# Cleaning data

## Dropping columns

### With more than 50% missing values

```python
df = df.loc[:, df.isnull().mean() < 0.5]
```

```python
df.info()
```

### Unnecessary columns

```python
columns_to_remove = [
    'id',
    'listing_url',
    'scrape_id',
    'picture_url',
    'host_id',
    'host_url',
    'host_name',
    'host_since',
    'host_location',
    'host_about',
    'host_has_profile_pic',
    'host_thumbnail_url',
    'host_picture_url',
    'host_has_profile_pic',
]
df.drop(columns_to_remove, axis=1, inplace=True)
df.head()
```
